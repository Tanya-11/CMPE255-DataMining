```python
#import all the necessary libraries

import pandas as pd
from collections import defaultdict
import re
import numpy as np
import ftfy
import datetime
import math
```

```python
#create functions for different tasks and add the description of each funtion. Below is the example

def load_data(filename):
    """Given a filename of a csv load data into a Pandas dataframe.

        filename - string

        return Pandas dataframe
    """
    missing_value_formats = ["n.a.","Na","NA","n/a", "na", "N/A","Null"]
    return pd.read_csv(filename, encoding="ISO-8859-1",na_values = missing_value_formats)
```

```python
# 1. Open the csv with your favorite spreadsheet application (excel, google sheets,
# numbers, etc.)

file_path = '/Users/Checkout/Downloads'

def main():
    """
    Clean up the Superhero dataset
    """
    
    # 2. Rename filename.csv to the data filename.
    superheros_df = load_data(file_path + "/Superheros_data.csv")

    # Q4 
    superheros_df.drop('Phase',axis=1 ,inplace=True);
   
    
    #Q5
    #I have removed one column called 'Phase' because I don't see it as a relevant attribute in the dataset. 
    #I can't predict what phases mean for this dataset.

    #Q6 
    print("Missing values for each column")
    for name, values in superheros_df.iteritems():
         print('{name}: {value}'.format(name=name, value=superheros_df[name].isna().sum()))  
    print("******************")
    
    #Q7
    superheros_df = superheros_df.fillna('')
    
    #Q8.a
    superheros_df['Filming Locations'] = superheros_df['Filming Locations'].str.upper();
    
    #Q8.b
    superheros_df['Filming Locations'] = superheros_df['Filming Locations'].str.replace('United Kingdom', 'UK', flags=re.I)
    superheros_df['Filming Locations'] = superheros_df['Filming Locations'].str.replace('United states', 'USA', flags=re.I)
    
    #Q9 
    print("Directors who have worked for both Marvel and DC-")
    directors = []
    combined = defaultdict(list)
    for index, row in superheros_df.iterrows():
        combined[row["Director"]].append(row["Company"])

    for director, company in combined.items():
        if "Marvel" in company and "DC" in company:
            directors.append(director)
            
    print(directors)
    
    #Q10
    for index, row in superheros_df.iterrows():
        if(type(row["Movie Run Time"]) == str):
            temp = row["Movie Run Time"].split(" ")[0]
            if(temp!=""):
                 if(int(temp)<10 or int(temp)>300):
                    temp=0
            row["Movie Run Time"].split(" ")[0] = temp
        superheros_df.loc[index, "Movie Run Time"] = temp       
    
    #Q11
    for index, row in superheros_df.iterrows():
        superheros_df.loc[index, "Title"] = ftfy.fix_text(row["Title"])
        superheros_df.loc[index, "Director"] = ftfy.fix_text(row["Director"])
       
    
    #Q12
    print("******************")
    print("Outlier for Review Rating:")
    for index, row in superheros_df.iterrows():
        if(type(row['Review Rating'])!= str):
            if(row['Review Rating'] < 1 or row['Review Rating'] > 10):
                print(row['Review Rating'])
    
    print("Outlier for Budget:")
    for index, row in superheros_df.iterrows():
        if(row['Budget'] != ""):
            if(int(row['Budget']) < 0):
                print(row['Budget'])
        else: print(row['Budget'])
    print("******************")

    #Q13
    sum = 0
    for index, row in superheros_df.iterrows():
        if(row['Status'] == "Released" and type(row['Review Rating']) != str):
            if(row['Review Rating'] > 0 and row['Review Rating'] <= 10):
                sum = sum + row['Review Rating']
    mean = round((sum/index),1)
    for index, row in superheros_df.iterrows():
        if(row['Status'] == "Released"):
             if(row['Review Rating'] < 1 or row['Review Rating'] > 10):
                    superheros_df.loc[index, "Review Rating"] = mean  
        else: superheros_df.loc[index, "Review Rating"] = ''
    
    #Q14
    superheros_df.drop(superheros_df[pd.DatetimeIndex(superheros_df['Release Date']).year < 2005].index, inplace = True)
    
    #Q15
    superheros_df.to_csv("output.csv")
    
if __name__ == "__main__":
    main()
```
